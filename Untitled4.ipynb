```
import math
def text_prob(model, text, delta=0):
    line = text
    line = re.sub(r"[*'*]"," ", line)
    words = line.split()
    arr = []
    '''    for i in range(model.n-1):
      arr.append('<s>')
    '''
    for word in words:
        word = word.lower()
        cont=re.sub(r'[,-?:!]'," ",word)
        cont = cont.strip()

        cont=re.sub(r'[^\w\s]', '', cont)

        tokens=re.sub(r'[0-9]+', ' ', cont)
        tokens=tokens.strip("\n")
        #print(repr(tokens))
        
        arr.append(tokens)
    ngrams = create_ngrams(arr, model.n)
    p=0.0
    for ng in ngrams:
        
        p += math.log(model.word_prob(ng[model.n-1],tuple(ng[:model.n-1]), delta))
    return p
```

```
def mask_rare(corpus):
  f = open(corpus, 'r')
  text = f.readlines()
  vocab = {}
  for line in text:

    line = re.sub(r"[*'*]"," ", line)
    words = line.split()
    for word in words:
        word = word.lower()
        cont=re.sub(r'[,-?:!]'," ",word)
        cont = cont.strip()

        cont=re.sub(r'[^\w\s]', '', cont)

        tokens=re.sub(r'[0-9]+', ' ', cont)
        #print(repr(tokens))
        if tokens in vocab:
            vocab[tokens] = vocab[tokens]+1
        else:
            vocab[tokens] = 1  
  rare_words = []
  for word in vocab:
    if vocab[word] == 1:
      rare_words.append(word)
  return rare_words
```

```
def replace_rare_words(corpus, rare_words):
  f = open(corpus, 'r')
  fw = open('new_corpus.txt','a')
  x = f.readlines()
  for line in x:
    line = re.sub(r"[*'*]"," ", line)
    line_bc = line
    words = line.split()
    for word in words:
        word_bc = word
        word = word.lower()
        cont=re.sub(r'[,-?:!]'," ",word)
        cont = cont.strip()

        cont=re.sub(r'[^\w\s]', '', cont)

        tokens=re.sub(r'[0-9]+', ' ', cont)
        if tokens in rare_words:
          fw.write(' <unk> ')
        else:
          fw.write(word_bc+" ")
    fw.write('\n')
```

```
class NGramInterpolator:
  def __init__(self, n, lambdas):
    self.n = n
    self.lambdas = lambdas
    self.lms = []
    bn = n
    for l in self.lambdas:
      ng= NGramLMUnk(bn)
      ngrs = ng.preprocess('new_corpus.txt')
      ng.get_other_dicts(ngrs)
      bn-=1
      self.lms.append(ng)
      
  def update(self, text):
    for lm in self.lms:
      lm.update(text)
  
  def word_prob(self, word, context, delta=0):
    pfinal = 0.0
    for i in range(len(self.lambdas)):
      pfinal += self.lms[i].word_prob(word, context, delta)
    return pfinal
   
```

```
def perplexity(model, corpus, delta=0):
  arr = preprocess(corpus)
  N = len(arr)
  f = open(corpus, 'r')
  x = f.readlines()
  logp = 0.0
  for line in x:
    logp += text_prob(model, line, delta)
  logp/=N
  return 2**(-1*logp)
```
